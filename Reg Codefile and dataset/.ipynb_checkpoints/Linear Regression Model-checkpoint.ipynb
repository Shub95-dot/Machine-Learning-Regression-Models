```python
# import important libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import joblib
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Load the dataset
data = pd.read_excel("ad_revenue_data.xlsx")
data.head()
```

```python
#Defininf independent and dependent variables
X = data[['AD']] #advertising spend as independent variable
Y = data['R'] #Sales revenue as dependent variable
```

```python
#splitting the data as train test split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
#Initialize the linear regression model
LR_model = LinearRegression()
```

```python
#Training the model
LR_model.fit(X_train,Y_train)
print("Model training Complete")
```

```python
#Getting the prediction from the trained model
y_pred = LR_model.predict(X_test)

mse = mean_squared_error(Y_test, y_pred)
r2 = r2_score(Y_test, y_pred)
print(f"Mean Squared Error: {mse}")
print(f"R2_score: {r2}")
```

```python
#Get the intercept and slope coefficients
intercept = LR_model.intercept_
slope= LR_model.coef_[0]

print(f"Intercept of this model is: {intercept}")
print(f"Slope of the model is: {slope}")
```

```python
#Plotting regression line
plt.scatter(X,Y, color='blue', label='Actual')
plt.plot(X_test,y_pred,color='red',linewidth=2,label='Predicted')

#adding equation to the plot
equation_text = f'R = {intercept:.2f} + {slope:.2f} * AD'
plt.text(X_test.max(), Y.min(), equation_text, fontsize=12, color='green', ha='right')

#Setting titles to the plot
plt.title('Simple Linear Regression : AD vs R')
plt.xlabel('Advertising Spend (AD)')
plt.ylabel('Sales Revenue (R)')
plt.legend()

plt.show()
```

```python
#Training the final model on the entire data set
#calling the model again
final_model = LinearRegression()

#Training on the entire dataset
final_model.fit(X,Y)
```

```python
#saving the model
joblib.dump(final_model, 'final_ad_revenue_model.pkl')
print('Final model trained on all data and saved!')
```
