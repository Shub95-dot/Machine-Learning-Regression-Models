```python
#import important libraries
import pandas as pd
import numpy as np
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
import joblib
```

I will not split the data as the other codefiles i have already split the data.
Its NECESSARY and MANDATORY to split the data or else you ML model will always overfit


```python
#Load the dataset
df = pd.read_excel("sample_data.xlsx")
df.head()
```

```python
#Define indepoendent and dependent variable
X = df[["AD","T","E"]]
y = df["R"] #revenue
```

```python
#initialise the SVR model
svr_model = SVR(kernel='rbf',C=1.0,epsilon=0.1) #all are default values
```

```python
#Training the model 
svr_model.fit(X,y)
```

```python
#Make predictions
y_pred = svr_model.predict(X)
```

```python
#Evaluate the model 

mse = mean_squared_error(y,y_pred)
r2 = r2_score(y,y_pred)
```

```python
#reporting evaluation metrics
print("Mean Squared Error:", mse)
print("R2 Score:", r2)

print("\nSVR model used with RBF kernel, C=1.0 and epsilon = 0.1.")
```
